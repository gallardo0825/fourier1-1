```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
t = np.arange(0,2*np.pi,0.01)
y = np.zeros(len(t))
for i in range(len(t)):
   if t[i] <= np.pi:
       y[i] = -1+2*t[i]/np.pi
   else:
       y[i] = 3-2*t[i]/np.pi
d = 360
theta = np.linspace(0,2*np.pi,d+1)
attempts = np.array([3,5,10])
for i in range(3):
   n = attempts[i]
   z = np.zeros(d+1)
   for k in range(1,n,2):
      ak = -8/np.pi**2/k**2
      z += ak*np.real(np.exp(1.j*k*theta))
   plt.plot(theta,z,label = "N = " + str(n))
plt.plot(t,y)
plt.legend(loc='upper right',fontsize=10)
```
