```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
t = np.arange(0,2*np.pi,0.01)
y = np.zeros(len(t))
for i in range(len(t)):
   if t[i] <= np.pi:
       y[i] = -1+2*t[i]/np.pi
   else:
       y[i] = 3-2*t[i]/np.pi
n = 360
theta = np.linspace(0,2*np.pi,n+1)
z = np.zeros(n+1)
for k in range(1,5,2):
   ak = -8/np.pi**2/k**2
   z += ak*np.real(np.exp(1.j*k*theta))
plt.plot(t,y)
plt.plot(theta,z)
```
