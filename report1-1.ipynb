```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
d = 360
theta = np.linspace(0,2*np.pi,d+1)
y = np.zeros(len(theta))
for i in range(len(theta)):
   if theta[i] <= np.pi:
       y[i] = -1+2*theta[i]/np.pi
   else:
       y[i] = 3-2*theta[i]/np.pi
attempts = np.array(range(1,101))
individualerror = np.zeros(len(theta))
error = np.zeros(len(attempts))
for i in range(len(attempts)):
   n = attempts[i]
   z = np.zeros(d+1)
   for k in range(1,n,2):
      ak = -8/np.pi**2/k**2
      z += ak*np.real(np.exp(1.j*k*theta))
   individualerror = abs(z-y)
   error[i] = sum(individualerror)
plt.plot(error)
plt.ylabel("the norm of error")
plt.xlabel("the nuber of attempts")
```
